# Mã giả & Lưu đồ giải thuật

### Lập trình

Lập trình là gì?

Lập trình là đưa ra những câu lệnh để máy tính thực hiện.
Đưa ra câu lệnh A trước câu lệnh B thì câu lệnh A sẽ được thực hiện trước.

### Thuật toán

Thuật toán là gì?

Thuật toán là những câu lệnh được đưa ra để giải quyết 1 vấn đề nào đó.

### Mã gỉả

#### Khái niệm:

Mã giả được viết với ngôn ngữ tự nhiên một thuật toán nào đó dùng để giải quyết một vấn đề lập trình.

Mã giả được quy ước chung theo từng nơi làm việc.

Ví dụ:

<b>Đề bài:</b>

Yêu cầu bài toán: Khi ta đói và thèm một món nào đó (nhập từ bàn phím), tiến hành đặt đồ ăn, nếu món ta thèm còn thì đặt đồ món đó, không còn thì đặt cơm sườn, không còn cơm sườn thì đi ngủ.

<b>Mã giả:</b>

Nhập (món A)
Nếu (món A còn):
    Ta đặt món A
Ngược lại:
    Nếu (cơm sườn còn)
        Ta đặt cơm sườn
    Ngược lại:
        Đi ngủ

<b>Đề bài:</b>

Yêu cầu bài toán: Nhập vào một số a, kiểm tra xem số đó có phải là số lớn hơn 10 không.

<b>Mã giả:</b>

Nhập ( a)
Nếu (a > 10):
    In ra (a là số lớn hơn 10)
Ngược lại:
    In ra (a là số bé hơn hoặc bằng 10)

<b>Đề bài:</b>

Yêu cầu bài toán: Nhập vào một số a, kiểm tra xem số a là số âm hay số dương

<b>Mã giả:</b>

Nhập ( a)
Nếu (a>0):
    In ra (a là số dương)
Ngược lại:
    Nếu (a==0):
        In ra (a là số không âm không dương)
    Nếu (a<0):
        In ra (a là số âm)

<b>Lưu đồ giải thuật:</b>

<b>Đề bài</b>

Nhập vào lương tháng này, nếu lương tháng này nhiều hơn 2 triệu thì trừ đi 10% thuế, nếu không thì không trừ.

Hãy tính và in ra lương thực nhận được

<b>Mã giả:</b>


Nhập (lương tháng này)
Nếu (lương tháng này > 2 triệu):
    lương thực nhận = lương tháng này - lương tháng này * 10%
    In ra (lương thực nhận)
Ngược lại:
    lương thực nhận = lương tháng này
    In ra (lương thực nhận)

<b>Lưu đồ giải thuật:</b>

<b>Đề bài</b>

Nhập vào lương thực nhận được tháng này, nếu lương ít hơn 10 triệu, đưa hết cho vợ, nếu lương nhiều hơn 10 triệu, vợ cho 10% lương

Hãy tính và in ra màn hình phần lương mình được nhận

<b>Mã giả:</b>

Nhập (lương thực nhận)
Nếu (lương thực nhận < 10 triệu):
    đưa hết cho vợ
    lương thực nhận = 0
    In ra (lương thực nhận)
Ngược lại
    Nếu (lương thực nhận > 10 triệu):
        vợ cho 1 triệu
        lương thực nhận = 10 triệu - 1 triệu
        In ra (lương thực nhận)
    Nếu (lương thực nhận == 10 triệu):
        cầm hết
        lương thực nhận = 10 triệu
        In ra (lương thực nhận)

<b>Mã giả:</b>

Nhập (điểm văn)
Nhập (điểm toán)
điểm trung bình = (điểm văn + điểm toán)/2
Nếu (điểm trung bình>8):
    In ra (Học sinh giỏi)

    In ra (Không phải học sinh giỏi)

<b>Mã giả:</b>

Nhập (số a)
Phần dư = a % 5
Nếu (Phần dư == 0):
    In ra (a chia hết cho 5)
Ngược lại:
    In ra (a không chia hết cho 5)

<b>Đề bài</b>

Nhập vào điểm số, nếu điểm số lớn hơn hoặc bằng 8, in ra dòng chữ Học sinh giỏi, nếu điểm số bé hơn 8 và lớn hơn hoặc bằng 6, in ra dòng chữ chữ học sinh Khá, ngược lại in ra dòng chữ Học sinh trung bình

<b>Mã giả:</b>

Nhập (điểm số)
Nếu (điểm số >= 8):
    In ra (Học sinh giỏi)
Nếu (6 <= điểm số < 8):
    In ra (Học sinh khá)
Ngược lại:
    In ra (Học sinh trung bình)

Nhập vào một số, kiểm tra tính chẳn lẻ và số đó có lớn hơn 10 hay không. (in ra 4 trường hợp: chẳn lớn hơn 10, lẻ lớn hơn 10, chẳn bé hơn hoặc bằng 10, lẻ bé hơn hoặc bằng 10)

Nhập (số a)
Phần dư = a % 2
Nếu (Phần dư == 0):
    Nếu a > 10:
        In ra (a là số chẵn lớn hơn 10)
    Nếu a < 10:
        In ra (a là số chẵn nhỏ hơn 10)
    Ngược lại:
        In ra (a = 10)
Ngược lại:
    Nếu a < 10:
        In ra (a là số lẻ nhỏ hơn 10)
    Nếu a > 10:
        In ra (a là số lẻ lớn hơn 10)
